# Challenge - Prime Locations

Given a 10x10x10 array of zeros, set (i,j,k) = 1 if i is odd, j is even, and k is prime.

In other words, set these elements to 1: (1,0,2), (1,0,3), (1,0,5), (1,0,7), (1,2,2), ...

```python
import numpy as np

chewy = np.zeros((10,10,10))

print(chewy)
```

```python
chewy[np.arange(1, chewy.shape[0], 2)[:, None, None], np.arange(0, chewy.shape[0], 2)[None, :, None], np.array([2,3,5,7])[None, None, :]] = 1
chewy
```
